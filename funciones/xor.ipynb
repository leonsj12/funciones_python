## Función `xor`

La función `xor` calcula el operador lógico XOR (exclusive OR) entre dos valores booleanos. El operador XOR devuelve `True` únicamente si uno de los dos valores es `True`, pero no ambos.

### Propósito

Esta función realiza una operación lógica XOR entre dos valores booleanos, utilizando una implementación manual del operador. Retorna el resultado de la operación como un valor booleano (`True` o `False`).

### Parámetros

 `a` (bool): Primer valor booleano que se desea evaluar.

 `b` (bool): Segundo valor booleano que se desea evaluar.

### Retorno

La función retorna un valor booleano:

 `True` si exactamente uno de los valores es `True`.

`False` si ambos valores son iguales (ambos `True` o ambos `False`).

### Implementación del Operador XOR

El operador XOR se puede implementar con la siguiente expresión lógica:
```python
xor = (a or b) and not (a and b)


````python
def xor(a, b):
    """
    Calcula el operador lógico XOR (exclusive OR) entre dos valores booleanos.

    Parámetros:
        a (bool): Primer valor booleano.
        b (bool): Segundo valor booleano.

    Retorna:
        bool: El resultado de a XOR b.
    """
    # Ejecutar la operación lógica XOR
    xor = (a or b) and not (a and b)  # Implementación lógica de XOR

    return xor

# Ejemplo de uso
a = True
b = False

resultado = xor(a, b)
print(f"El resultado de {a} XOR {b} es {resultado}")
````

Esto asegura que el resultado es True si exactamente uno de los valores es True, y False en caso contrario.

**Ojo:**

La función maneja exclusivamente valores booleanos (True o False).
Si se proporciona un valor no booleano, la función podría producir un comportamiento inesperado o un error.
